```python
import torch
import matplotlib.pyplot as plt
```

```python
dist1 = torch.distributions.Normal(0,1)
dist2 = torch.distributions.Normal(10, 5)

N = 10000
x1 = dist1.rsample((N,))
x2 = dist2.rsample((N,))

x3 = 10 + 5*x1
```

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.hist(x1, bins=20)
plt.show()
plt.close()
```

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.hist(x2, bins=20)
ax.hist(x3, bins=20)
plt.show()
plt.close()
```
